# Código

## Importación e instalación de librerías

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from funpymodeling.exploratory import status
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import accuracy_score
```

## Configuración del display de dataframes

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)
```

## Importación y análisis inicial de datos

```python
data=pd.read_csv('heart_2020_cleaned.csv', sep=',')
```

```python
data.head(5)
```

```python
data.describe(include="all")
```

```python
status(data)
```

```python
data.info()
```

## Transformación de variables categóricas en binarias

```python
class_map = {'Yes':1, 'No':0}
boolean_columns = ['HeartDisease', 'Smoking', 'AlcoholDrinking', 'Stroke', 'DiffWalking', 'PhysicalActivity', 'Asthma', 'KidneyDisease', 'SkinCancer', 'Diabetic']
for column in boolean_columns:
    data[column] = data[column].map(class_map)
```

```python
class_map = {'Female':1, 'Male':0}
data['Sex'] = data['Sex'].map(class_map)
```

### Extra: transformación de otras variables categóricas que se pidieron quitar

```python
# Obtención de todos los posibles valores de la variable GenHealth para transformarlos en numéricos
data.GenHealth.unique()
```

```python
# Mantenemos un orden: los valores con "mejor" valor corresponden a un número más alto 
class_map = {'Poor':1, 'Fair':2, 'Good':3, 'Very good':4, 'Excellent':5}
data['GenHealth'] = data['GenHealth'].map(class_map)
```

```python
class_map = {'White':1, 'Black':2, 'Asian':3, 'American Indian/Alaskan Native':4, 'Hispanic':5, 'Other':6}
data['Race'] = data['Race'].map(class_map)
```

```python
# Mantenemos un orden de rangos etarios ascendentes
class_map = {'18-24':1, '25-29':2, '30-34':3, '35-39':4, '40-44':5, '45-49':6, '50-54':7, '55-59':8, '60-64':9, '65-69':10, '70-74':11, '75-79':12, '80 or older':13}
data['AgeCategory'] = data['AgeCategory'].map(class_map)
```

### Revisión de los valores convertidos

```python
data.sample(3)
```

```python
data.dtypes
```

## Limpieza de datos

```python
# Quitamos las variables solicitadas
# A modo informativo, dejamos comentados los scores obtenidos al mantener columnas que se pidieron quitar

data_filtered = data.drop(['AgeCategory', 'Race', 'Diabetic', 'GenHealth'], axis=1) # 0.9790647133319783
# data_filtered = data.drop(['AgeCategory', 'Race'], axis=1) # 0.9883050078956832
# data_filtered = data.drop('Race', axis=1) # 0.9960013446113917
# data_filtered = data # 0.9969757344006862
```

```python
# Separamos la variable target de las variables predictoras
x_data = data_filtered.drop('HeartDisease', axis=1)
y_data = data_filtered['HeartDisease']
```

## Creación del modelo y obtención de resultados

```python
# Divimos los datos en conjuntos de train y test
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=42)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
# Creación y entrenamiento del modelo
rf = RandomForestClassifier(n_estimators=100, random_state=42)

rf.fit(x_train, y_train)
```

```python
# Ejecutamos la predicción del modelo en nuestros conjuntos de datos

# En training
pred_tr=rf.predict(x_train)

# En testing
pred_ts=rf.predict(x_test)
```

### Evaluación de los datos de entrenamiento

```python
# El dataframe df_val_tr guardará los valores originales de la variable target para cada registro de test,
# y su correspondiente predicción del modelo
df_val_tr=pd.DataFrame({'y_train':y_train, 'pred_tr':pred_tr})
```

```python
df_val_tr
```

```python
# Obtenemos la proporción de aciertos de manera manual
# es lo mismo que accuracy_score(df_val_tr.y_train, df_val_tr.pred_tr, normalize=True)
sum(df_val_tr.y_train==df_val_tr.pred_tr)/len(df_val_tr)
```

```python
print("Accuracy score obtenido en datos de entrenamiento:")
accuracy_score_with_100_estimators_tr = accuracy_score(df_val_tr.y_train, df_val_tr.pred_tr, normalize=True)
print(accuracy_score_with_100_estimators_tr)
```

### Evaluación de los datos de test

```python
df_val_ts=pd.DataFrame({'y_test':y_test, 'pred_ts':pred_ts})
```

```python
print("Accuracy score obtenido en datos de test:")
accuracy_score_with_100_estimators_ts = accuracy_score(df_val_ts.y_test, df_val_ts.pred_ts, normalize=True)
print(accuracy_score_with_100_estimators_ts)
```

## Variación de hiperparámetros 

### n_estimators = 500  y  n_jobs = -1

- n_estimators es la cantidad de árboles de decisión que conformarán el random forest
- n_jobs permite seleccionar la cantidad de procesadores a utilizar para ejecutar el modelo. Un valor de **-1** permite utilizar todos los procesadores disponibles (como se pide en la consigna)

```python
rf_1 = RandomForestClassifier(n_estimators=500, random_state=42, n_jobs=-1)

rf_1.fit(x_train, y_train)
```

#### Evaluación de los datos de entrenamiento

```python
pred_tr_1=rf_1.predict(x_train)
```

```python
df_val_tr_1=pd.DataFrame({'y_train':y_train, 'pred_ts':pred_tr_1})
```

```python
print("Accuracy score obtenido en datos de training:")
accuracy_score_with_500_estimators_tr = accuracy_score(df_val_tr_1.y_train, df_val_tr_1.pred_ts, normalize=True)
print(accuracy_score_with_500_estimators_tr)
```

#### Evaluación de los datos de test

```python
pred_ts_1=rf_1.predict(x_test)
```

```python
df_val_ts_1=pd.DataFrame({'y_test':y_test, 'pred_ts':pred_ts_1})
```

```python
print("Accuracy score obtenido en datos de test:")
accuracy_score_with_500_estimators_ts = accuracy_score(df_val_ts_1.y_test, df_val_ts_1.pred_ts, normalize=True)
print(accuracy_score_with_500_estimators_ts)
```

### n_estimators = 5

```python
rf_2 = RandomForestClassifier(n_estimators=5, random_state=42)

rf_2.fit(x_train, y_train)
```

#### Evaluación de los datos de entrenamiento

```python
pred_tr_2=rf_2.predict(x_train)
```

```python
df_val_tr_2=pd.DataFrame({'y_train':y_train, 'pred_tr':pred_tr_2})
```

```python
print("Accuracy score obtenido en datos de entrenamiento:")
accuracy_score_with_5_estimators_tr = accuracy_score(df_val_tr_2.y_train, df_val_tr_2.pred_tr, normalize=True)
print(accuracy_score_with_5_estimators_tr)
```

#### Evaluación de los datos de test

```python
pred_ts_2=rf_2.predict(x_test)
```

```python
df_val_ts_2=pd.DataFrame({'y_test':y_test, 'pred_ts':pred_ts_2})
```

```python
print("Accuracy score obtenido en datos de test:")
accuracy_score_with_5_estimators_ts = accuracy_score(df_val_ts_2.y_test, df_val_ts_2.pred_ts, normalize=True)
print(accuracy_score_with_5_estimators_ts)
```

# Conclusiones

Analizando los resultados obtenidos:

```python
print("Accuracy score 100 estimadores")
print(f"Training: {accuracy_score_with_100_estimators_tr}")
print(f"Test: {accuracy_score_with_100_estimators_ts}")
print("")
print("Accuracy score 500 estimadores")
print(f"Training: {accuracy_score_with_500_estimators_tr}")
print(f"Test: {accuracy_score_with_500_estimators_ts}")
print("")
print("Accuracy score 5 estimadores")
print(f"Training: {accuracy_score_with_5_estimators_tr}")
print(f"Test: {accuracy_score_with_5_estimators_ts}")
```

Podemos concluir que:
- Utilizando el mismo tipo de modelo (RandomForestClasifier), utilizando el mismo random_state y variando el parámetro n_estimators, el accuracy score varía levemente. Por lo tanto, si hay que evaluar el coste computacional, creemos que no es necesario ejecutar el modelo con 500 estimators, ya que su diferencia en puntuación es insignificantemente superior.
- El parámetro n_jobs no influye en el score, solo influye en la velocidad de procesamiento del modelo.
- En este ejercicio hubiese sido útil puntuar los modelos de otra forma. Hay que darle más importancia en la evaluación a los falsos negativos, ya que es el caso que más queremos evitar: diagnosticar que una persona NO tendrá un ataque al corazón cuando SI podría tenerlo.
