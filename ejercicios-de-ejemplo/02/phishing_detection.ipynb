## 3) Acondicionar dataset ðŸ”§

**Eliminar columna url**
<br> No tiene un patrÃ³n, no nos aporta al anÃ¡lisis.

````python
filter_data = data.drop(['url'], axis=1)
````

Si se desea eliminar mÃ¡s columnas, lo que se hace es aÃ±adirlas todas dentro de una lista [].
<br> Por ejemplo:
<br> filter_data = data.drop(['url','nb_or','ratio_nullHyperlinks], axis=1)


**Valores categÃ³ricos a numÃ©ricos**
<br> Si la url es legÃ­tima le asignamos un 0, caso contrario un 1.

````python
filter_data.status.value_counts()
````

````python
class_map = {'legitimate':0, 'phishing':1}
filter_data['status'] = filter_data['status'].map(class_map)
````

````python
filter_data.head(5)
````

## 4) SeparaciÃ³n input/output o X/Y âœ‚ï¸

````python
x_data=filter_data.drop('status', axis=1)
````

````python
y_data=filter_data['status']
````

*Nota:* Es comÃºn ver el "dropeo" de columnas con slicing/iloc. 
<br>No lo aconsejo porque no somos explicitos con la columna que estamos eliminando. 
<br>AdemÃ¡s estarÃ­amos asumiendo que la variable a predecir esta al final, y a veces no ocurre.

Ej:

`x_data=data.iloc[:, :-1]`

`y_data=data.iloc[:, -1]`

## 5) SeparaciÃ³n training / test ðŸ’ª

````python
#!pip insta scikit-learn
````

````python
from sklearn.model_selection import train_test_split 
````

70% Training - 30% Testing
70% Training - 15% Validation - 15% Testing   

````python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=42)

print(x_train.shape, y_train.shape)
````

````python
print(x_test.shape, y_test.shape)
````

````python
x_train.head(2)
````

## 6) CreaciÃ³n del modelo: Random Forest ðŸ˜Ž

````python
# Importamos el modelo que queremos
from sklearn.ensemble import RandomForestClassifier 
````

````python
# Creamos 100 decision trees
rf = RandomForestClassifier(n_estimators=100)

rf.fit(x_train, y_train)
````

## 7) Predecimos la clase âœ…

````python
# En training
pred_tr=rf.predict(x_train)

# En testing
pred_ts=rf.predict(x_test)
````

````python
pred_ts
````

### 7.1 Validamos Training
Generamos un dataframe con la realidad vrs. lo predicho

````python
df_val_tr=pd.DataFrame({'y_train':y_train, 'pred_tr':pred_tr})

# Â¿cuÃ¡ntos aciertos?
sum(df_val_tr.y_train==df_val_tr.pred_tr)/len(df_val_tr)
````

````python
df_val_tr
````

````python
from sklearn.metrics import accuracy_score
````

````python
accuracy_score(df_val_tr.y_train, df_val_tr.pred_tr, normalize=True)
````

### 7.2 Validamos Testing
Generamos un dataframe con la realidad vrs. lo predicho

````python
df_val_ts=pd.DataFrame({'y_test':y_test, 'pred_ts':pred_ts})
````

````python
# Â¿cuÃ¡ntos aciertos?
accuracy_score(df_val_ts.y_test, df_val_ts.pred_ts, normalize=True)
````
