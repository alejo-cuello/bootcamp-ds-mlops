```python
import pickle
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
```

```python
with open('data_norm.pickle', 'rb') as handle:
     data_norm = pickle.load(handle)
```

```python
data_norm
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(data_norm)

# --- Step 3: Apply t-SNE to reduce to 2D ---
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, random_state=42)
X_tsne = tsne.fit_transform(data_norm)
```

```python
import pandas as pd
from matplotlib import pyplot as plt
tsne_df = pd.DataFrame(X_tsne, columns=["TSNE1", "TSNE2"])
tsne_df["cluster"] = labels

# --- Step 5: Visualize ---
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=tsne_df, x="TSNE1", y="TSNE2",
    hue="cluster", palette="Set1", s=100, alpha=0.7
)
plt.title("t-SNE visualization of KMeans clusters")
plt.show()
```
